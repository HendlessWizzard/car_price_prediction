```python
def col_display(text, color="black", bold=False, bgcolor=None):
    # –î–æ–±–∞–≤–ª—è–µ–º –ø—Ä–æ–±–µ–ª –≤ –Ω–∞—á–∞–ª–µ —Ç–µ–∫—Å—Ç–∞
    text = "'" + text

    # –°—Ç–∏–ª—å –¥–ª—è —Ç–µ–∫—Å—Ç–∞
    style = f"color: {color};"
    
    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∂–∏—Ä–Ω–æ—Å—Ç–∏, –µ—Å–ª–∏ –Ω—É–∂–Ω–æ
    if bold:
        style += " font-weight: bold;"
    
    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Ñ–æ–Ω–∞, –µ—Å–ª–∏ —É–∫–∞–∑–∞–Ω
    if bgcolor:
        style += f" background-color: {bgcolor};"
    
    # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º margin –∏ padding, —á—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å –æ–±—Ä–µ–∑–∫–∏ —Ç–µ–∫—Å—Ç–∞
    style += " padding: 0; margin: 0; display: inline;"

    # –§–æ—Ä–º–∏—Ä—É–µ–º HTML –¥–ª—è –≤—ã–≤–æ–¥–∞
    html_text = f"<div style='{style}'>{text}</div>"
    
    # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º —Ç–µ–∫—Å—Ç
    display(HTML(html_text))
    
def err_display(text):
    col_display('ERROR: ' + text, color='white', bgcolor="red", bold=True)
    
def info_display(text):
    col_display('INFO: ' +  text, color='white', bgcolor="blue", bold=True)
    
def cycle_display(text):
    col_display('–¢–µ–∫—É—â–∞—è –∏—Ç–µ—Ä–∞—Ü–∏—è: ' +  text, color='yellow', bold=True)
    
def final_display(text):
    col_display(text, color='green', bold=True)    
    
def title_display(text):
    col_display(text, color='white', bgcolor="green", bold=True)
```

# –û—Å–Ω–æ–≤–Ω—ã–µ —ç—Ç–∞–ø—ã –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ–π –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö

```python
# –ü—É—Ç–∏ –∫ —Ñ–∞–π–ª–∞–º
local_paths = {
    "autos": "autos.csv",
}

server_paths = {
    "autos": "https://code.s3.yandex.net/datasets/autos.csv",
}

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–≥—Ä—É–∑–∫–∏ –¥–∞–Ω–Ω—ã—Ö —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π –ø—É—Ç–∏
def load_data_safely(local_path, server_path, sep=',', decimal='.', encoding='utf-8'):
    """
    –ó–∞–≥—Ä—É–∂–∞–µ—Ç –¥–∞–Ω–Ω—ã–µ –∏–∑ —Ñ–∞–π–ª–∞, –ø—Ä–æ–≤–µ—Ä—è—è –Ω–∞–ª–∏—á–∏–µ –ª–æ–∫–∞–ª—å–Ω–æ–≥–æ –ø—É—Ç–∏ –∏–ª–∏ —Å–µ—Ä–≤–µ—Ä–Ω–æ–≥–æ.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        local_path (str): –õ–æ–∫–∞–ª—å–Ω—ã–π –ø—É—Ç—å –∫ —Ñ–∞–π–ª—É
        server_path (str): –°–µ—Ä–≤–µ—Ä–Ω—ã–π –ø—É—Ç—å –∫ —Ñ–∞–π–ª—É
        sep (str): –†–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å –∑–Ω–∞—á–µ–Ω–∏–π
        decimal (str): –†–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å –¥–µ—Å—è—Ç–∏—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
        encoding (str): –ö–æ–¥–∏—Ä–æ–≤–∫–∞ —Ñ–∞–π–ª–∞
    
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        pd.DataFrame: –ó–∞–≥—Ä—É–∂–µ–Ω–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
    """
    try:
        if os.path.exists(local_path):
            data = pd.read_csv(local_path, sep=sep, decimal=decimal, encoding=encoding)
            print(f"‚úÖ –£—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω —Ñ–∞–π–ª –∏–∑ –ª–æ–∫–∞–ª—å–Ω–æ–≥–æ –ø—É—Ç–∏: {local_path}")
        else:
            data = pd.read_csv(server_path, sep=sep, decimal=decimal, encoding=encoding)
            print(f"üåê –£—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω —Ñ–∞–π–ª —Å —Å–µ—Ä–≤–µ—Ä–∞: {server_path}")
        return data
    except FileNotFoundError:
        print(f"‚ùå –û—à–∏–±–∫–∞: –§–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω - {local_path} –∏–ª–∏ {server_path}")
        return None
    except pd.errors.ParserError:
        print(f"‚ö†Ô∏è –û—à–∏–±–∫–∞: –ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç –¥–∞–Ω–Ω—ã—Ö - {local_path} –∏–ª–∏ {server_path}")
        return None
    except Exception as e:
        print(f"‚ùó –ù–µ–∏–∑–≤–µ—Å—Ç–Ω–∞—è –æ—à–∏–±–∫–∞ –ø—Ä–∏ –∑–∞–≥—Ä—É–∑–∫–µ {local_path} –∏–ª–∏ {server_path}: {e}")
        return None

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
autos = load_data_safely(local_paths["autos"], server_paths["autos"])

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –∏ –æ–±–∑–æ—Ä –¥–∞–Ω–Ω—ã—Ö
def overview_data(data, name):
    """
    –û—Ç–æ–±—Ä–∞–∂–∞–µ—Ç –∫—Ä–∞—Ç–∫—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        data (pd.DataFrame): –î–∞—Ç–∞—Ñ—Ä–µ–π–º –¥–ª—è –æ–±–∑–æ—Ä–∞
        name (str): –ò–º—è –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö
    """
    if data is not None:
        print(f"\nüìä –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö: {name}")
        display(data.head())
        data.info()
    else:
        print(f"‚ùå –î–∞–Ω–Ω—ã–µ {name} –Ω–µ –±—ã–ª–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã –∏–∑-–∑–∞ –æ—à–∏–±–æ–∫.")

# –í—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –¥–∞—Ç–∞—Å–µ—Ç–µ
if autos is not None:
    overview_data(autos, "autos")
else:
    print("‚ùå –û—à–∏–±–∫–∞: –î–∞—Ç–∞—Å–µ—Ç –Ω–µ –∑–∞–≥—Ä—É–∂–µ–Ω, –∞–Ω–∞–ª–∏–∑ –Ω–µ–≤–æ–∑–º–æ–∂–µ–Ω.")
```

**–ò–∑–º–µ–Ω–∏–ª–∏ –Ω–∞–∑–≤–∞–Ω–∏—è –∏—Å—Ö–æ–¥–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –Ω–∞ –±–æ–ª–µ–µ —É–¥–æ–±–Ω—ã–µ –∏ –≤—ã–≤–µ–ª–∏ –∫–ª—é—á–µ–≤—É—é –∏–Ω–æ—Ñ—Ä–º–∞—Ü–∏—é –ø–æ –ø—Ä–æ–ø—É—Å–∫–∞–º –∏ –¥—É–±–ª–∏–∫–∞—Ç–∞–º –≤ –∏—Å—Ö–æ–¥–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ**

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤
def rename_columns(autos):
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤:
    - –î–æ–±–∞–≤–ª—è–µ—Ç "_" –ø–µ—Ä–µ–¥ –∫–∞–∂–¥–æ–π –∑–∞–≥–ª–∞–≤–Ω–æ–π –±—É–∫–≤–æ–π (–∫—Ä–æ–º–µ –ø–µ—Ä–≤–æ–π).
    - –ü–µ—Ä–µ–≤–æ–¥–∏—Ç –Ω–∞–∑–≤–∞–Ω–∏—è –≤ –Ω–∏–∂–Ω–∏–π —Ä–µ–≥–∏—Å—Ç—Ä.
    """
    def convert_column_name(name):
        name = re.sub(r'(?<!^)(?=[A-Z])', '_', name)  # –î–æ–±–∞–≤–ª—è–µ–º "_" –ø–µ—Ä–µ–¥ –∑–∞–≥–ª–∞–≤–Ω—ã–º–∏ –±—É–∫–≤–∞–º–∏ (–∫—Ä–æ–º–µ –ø–µ—Ä–≤–æ–π)
        return name.lower()  # –ü—Ä–∏–≤–æ–¥–∏–º –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É

    autos.columns = [convert_column_name(col) for col in autos.columns]
    return autos
    
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
def handle_missing_data(autos, dataset_name):
    """
    –í—ã–≤–æ–¥–∏—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ –∏ –æ–±—â—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é.
    """
    missing_data = autos.isnull().sum()
    total_missing = missing_data.sum()

    title_display(f"\n–ü—Ä–æ–ø—É—Å–∫–∏ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ {dataset_name}:")
    if total_missing == 0:
        final_display("–ü—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ—Ç.")
    else:
        final_display("–ù–∞–π–¥–µ–Ω—ã –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:")
        display(missing_data[missing_data > 0])

    return autos

# –§—É–Ω–∫—Ü–∏—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞ –¥—É–±–ª–∏–∫–∞—Ç—ã
def check_duplicates(autos, dataset_name):
    """
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç –∏ –≤—ã–≤–æ–¥–∏—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –¥—É–±–ª–∏–∫–∞—Ç–∞—Ö.
    """
    duplicates = autos.duplicated().sum()
    title_display(f"\n–ü—Ä–æ–≤–µ—Ä–∫–∞ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ {dataset_name}:")
    if duplicates == 0:
        final_display("–î—É–±–ª–∏–∫–∞—Ç—ã –Ω–µ –Ω–∞–π–¥–µ–Ω—ã.")
    else:
        final_display(f"–ù–∞–π–¥–µ–Ω–æ {duplicates} –¥—É–±–ª–∏–∫–∞—Ç–æ–≤.")

# –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö
def preprocess_data(autos, dataset_name):
    """
    –ü—Ä–µ–¥–æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –¥–∞–Ω–Ω—ã–µ:
    - –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤.
    - –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ.
    - –ü—Ä–æ–≤–µ—Ä—è–µ—Ç –Ω–∞–ª–∏—á–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤.
    """
    title_display(f"\n–û–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö: {dataset_name}")
    
    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤
    autos = rename_columns(autos)
    title_display(f"–ù–æ–≤—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤:\n{autos.columns.tolist()}")
    
    # –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    autos = handle_missing_data(autos, dataset_name)

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –¥—É–±–ª–∏–∫–∞—Ç—ã
    check_duplicates(autos, dataset_name)

    return autos


# –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏
if autos is not None:
    autos = preprocess_data(autos, "autos")

    # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ø–µ—Ä–≤—ã—Ö —Å—Ç—Ä–æ–∫ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞
    title_display("\nüìä –ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:")
    display(autos.head())
else:
    err_display("–û—à–∏–±–∫–∞: –î–∞—Ç–∞—Å–µ—Ç –Ω–µ –∑–∞–≥—Ä—É–∂–µ–Ω, –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–∞.")
```

* **–ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –ø—Ä–æ–≤–µ—Ä–∫–∏ –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –æ–±–Ω–∞—Ä—É–∂–∏–ª–∏, —á—Ç–æ:**
    - –í –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö –∏—Å—Ö–æ–¥–Ω—ã –∏—Å—Ö–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã —Å–æ–¥–µ—Ä–∂–∞—Ç –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ 
    - –í –∏—Å—Ö–æ–¥–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—Ç –¥—É–±–ª–∏–∫–∞—Ç—ã
      
**–≠—Ç–∏ –¥–µ—Ñ–µ–∫—Ç—ã –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –±—É–¥—É—Ç –∏—Å–ø—Ä–∞–≤–ª–µ–Ω—ã –Ω–∞ —ç—Ç–∞–ø–µ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ –ø–∞–π–ø–ª–∞–π–Ω–∞(–ù–∞ –¥–∞–Ω–Ω–æ–º —ç—Ç–∞–ø–µ –ø—Ä–µ–¥–æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –¥–∞—Ç–∞—Å–µ—Ç –Ω–µ–ª—å–∑—è, –ø–æ–∫–∞ –Ω–µ –ø—Ä–æ–∏–∑–æ—à–ª–æ –µ–≥–æ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏. –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–∞, —Ç.–∫ –æ–Ω–∞ –¥–æ–ª–∂–Ω–∞ –∏–º–∏—Ç–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∞–ª—å–Ω—ã–π –Ω–µ–æ–±—Ä–∞–±–æ—Ç–Ω–Ω—ã–π –ø–æ—Ç–æ–∫ –¥–∞–Ω–Ω—ã—Ö)**

**–ü—Ä–æ–≤–µ—Ä–∏–ª–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –Ω–µ—è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤(–ò—Ö –∏–º–µ–µ—Ç —Å–º—ã—Å–ª –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∞—Ç—å –∑–∞—Ä–∞–Ω–µ–µ) –¢–∞–∫–∂–µ –ø—Ä–∏–≤–µ–ª–∏ —Å—Ç–æ–ª–±—Ü—ã –∫ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–º —Ç–∏–ø–∞–º –¥–∞–Ω–Ω—ã—Ö**

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∏–∑—É—á–µ–Ω–∏—è –Ω–µ—è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –∏ –∏—Ö –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–∏—è
def find_and_fix_implicit_duplicates(df, column_name):
    """
    –ò—â–µ—Ç –∏ –∏—Å–ø—Ä–∞–≤–ª—è–µ—Ç –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–º —Å—Ç–æ–ª–±—Ü–µ.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        df (pd.DataFrame): –î–∞—Ç–∞—Ñ—Ä–µ–π–º –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞.
        column_name (str): –ù–∞–∑–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞, –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω—É–∂–Ω–æ –∏—Å–∫–∞—Ç—å –¥—É–±–ª–∏–∫–∞—Ç—ã.

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        pd.DataFrame: –î–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏.
    """
    title_display(f"\nüîç –ê–Ω–∞–ª–∏–∑ –Ω–µ—è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–µ: {column_name}")
    
    unique_values_before = df[column_name].astype(str).str.lower().str.strip().value_counts()
    print(f"üìä –£–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (–¥–æ –æ–±—Ä–∞–±–æ—Ç–∫–∏):\n{unique_values_before}")

    # –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ –≤—Å–µ—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É –∏ —É–¥–∞–ª–µ–Ω–∏–µ –ª–∏—à–Ω–∏—Ö –ø—Ä–æ–±–µ–ª–æ–≤
    df[column_name] = df[column_name].astype(str).str.lower().str.strip()

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø–æ—Å–ª–µ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–∏—è
    unique_values_after = df[column_name].value_counts()
    print(f"\n‚úÖ –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (–ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏):\n{unique_values_after}")

    return df

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–∏–≤–µ–¥–µ–Ω–∏—è —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö
def convert_data_types(df):
    """
    –ü—Ä–∏–≤–æ–¥–∏—Ç —Å—Ç–æ–ª–±—Ü—ã –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –∫ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–º —Ç–∏–ø–∞–º –¥–∞–Ω–Ω—ã—Ö.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        df (pd.DataFrame): –î–∞—Ç–∞—Ñ—Ä–µ–π–º –¥–ª—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è.

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        pd.DataFrame: –î–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–Ω—ã—Ö.
    """
    print("\nüîÑ –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö...")

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –¥–∞—Ç –≤ datetime
    date_columns = ['dateCrawled', 'dateCreated', 'lastSeen']
    for col in date_columns:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors='coerce')
            display(f"üìÖ {col} ‚Üí datetime")

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
    numeric_columns = ['price', 'registration_year', 'power', 'kilometer']
    for col in numeric_columns:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            display(f"üî¢ {col} ‚Üí numeric")

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    categorical_columns = ['vehicle_type', 'gearbox', 'model', 'fuel_type', 'brand', 'repaired', 'number_of_pictures', 'postal_code', 'registration_month']
    for col in categorical_columns:
        if col in df.columns:
            df[col] = df[col].astype('category')
            display(f"üè∑ {col} ‚Üí category")

    display("‚úÖ –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö –∑–∞–≤–µ—Ä—à–µ–Ω–æ.")
    return df

# –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π
if autos is not None:
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º –∏ –∏—Å–ø—Ä–∞–≤–ª—è–µ–º –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
    for col in ['vehicleType', 'gearbox', 'fuelType', 'brand', 'repaired', 'model']:
        if col in autos.columns:
            autos = find_and_fix_implicit_duplicates(autos, col)

# –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö
autos = convert_data_types(autos)
```

**–ù–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –≤ –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç(–ù–∞ –≤—Å—è–∫–∏–π —Å–ª—É—á–∞–π –ø—Ä–∏–≤–µ–ª–∏ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É). –¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö –ø—Ä–∏–≤–µ–ª–∏ –∫ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–º. –û–¥–Ω–∞–∫–æ –≤ —Å—Ç–æ–ª–±—Ü–µ —Å —Ç–∏–ø–∞–º–∏ —Ç–æ–ø–ª–∏–≤–∞ –µ—Å—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è `petrol` –∏ `gasoline`(–û–Ω–∏ –æ–±–∞ –æ–∑–Ω–∞—á–∞—é—Ç –±–µ–Ω–∑–∏–Ω, –Ω–æ –æ–¥–∏–Ω –≤–∞—Ä–∏–∞–Ω—Ç –∞–Ω–≥–ª–∏–π—Å–∫–∏–π, –¥—Ä—É–≥–æ–π - –∞–º–µ—Ä–∏–∫–∞–Ω—Å–∫–∏–π). –ò–º–µ–µ—Ç —Å–º—ã—Å–ª –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —ç—Ç–∏—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –æ–¥–∏–Ω. –í —ç—Ç–æ–º –∂–µ —Å—Ç–æ–ª–±—Ü–µ –µ—Å—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ `other`, –∫–æ—Ç–æ—Ä–æ–µ –≤–µ—Ä–æ—è—Ç–Ω–æ, –æ–∑–Ω–∞—á–∞–µ—Ç —Ç–æ –∂–µ —Å–∞–º–æ–µ, —á—Ç–æ –∏ `electric`. –ò—Ö —Ç–æ–∂–µ —Å—Ç–æ–∏—Ç –æ–±—ä–µ–¥–∏–Ω–∏—Ç—å**

```python
# üîπ –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ 'fuel_type'
fuel_type_mapping = {
    'petrol': 'gasoline',  # –û–±—ä–µ–¥–∏–Ω—è–µ–º 'petrol' –∏ 'gasoline' –≤ –æ–¥–∏–Ω –∫–ª–∞—Å—Å
    'other': 'electric'  # –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º, —á—Ç–æ 'other' –æ–∑–Ω–∞—á–∞–µ—Ç 'electric'
}

# –ü—Ä–∏–º–µ–Ω—è–µ–º –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫–æ –≤—Å–µ–π —Ç–∞–±–ª–∏—Ü–µ data
autos['fuel_type'] = autos['fuel_type'].replace(fuel_type_mapping)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
autos['fuel_type'].value_counts()
```

### üîπ 1. –ò–∑–º–µ–Ω–µ–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤  
–î–ª—è —É–¥–æ–±—Å—Ç–≤–∞ —Ä–∞–±–æ—Ç—ã —Å –¥–∞–Ω–Ω—ã–º–∏ –∏—Å—Ö–æ–¥–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –±—ã–ª–∏ –ø—Ä–∏–≤–µ–¥–µ–Ω—ã –∫ –±–æ–ª–µ–µ —á–∏—Ç–∞–µ–º–æ–º—É –∏ —É–Ω–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω–æ–º—É —Ñ–æ—Ä–º–∞—Ç—É:
- –í—Å–µ –Ω–∞–∑–≤–∞–Ω–∏—è –ø–µ—Ä–µ–≤–µ–¥–µ–Ω—ã –≤ **–Ω–∏–∂–Ω–∏–π —Ä–µ–≥–∏—Å—Ç—Ä**.
- –î–æ–±–∞–≤–ª–µ–Ω—ã **`_` –º–µ–∂–¥—É —Å–ª–æ–≤–∞–º–∏** –¥–ª—è —É–¥–æ–±–Ω–æ–≥–æ –æ–±—Ä–∞—â–µ–Ω–∏—è –≤ –∫–æ–¥–µ.
- –£–¥–∞–ª–µ–Ω—ã **–ª–∏—à–Ω–∏–µ –ø—Ä–æ–±–µ–ª—ã –∏ —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã**.

### üîπ 2. –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –ø—Ä–æ–ø—É—Å–∫–∏ –∏ –¥—É–±–ª–∏–∫–∞—Ç—ã  
–ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –∞–Ω–∞–ª–∏–∑–∞ –æ–±–Ω–∞—Ä—É–∂–µ–Ω—ã —Å–ª–µ–¥—É—é—â–∏–µ –ø—Ä–æ–±–ª–µ–º—ã:
- –í **–Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö** –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—Ç **–ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è**. –≠—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å —Å–≤—è–∑–∞–Ω–æ —Å —Ç–µ–º, —á—Ç–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –Ω–µ –∑–∞–ø–æ–ª–Ω—è–ª–∏ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã–µ –ø–æ–ª—è –ø—Ä–∏ –ø–æ–¥–∞—á–µ –æ–±—ä—è–≤–ª–µ–Ω–∏—è.
- –í –¥–∞—Ç–∞—Å–µ—Ç–µ –Ω–∞–π–¥–µ–Ω—ã **—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã —Å—Ç—Ä–æ–∫**.
- –í—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è **–Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã** –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö (–Ω–∞–ø—Ä–∏–º–µ—Ä, —Ä–∞–∑–ª–∏—á–Ω—ã–µ –Ω–∞–ø–∏—Å–∞–Ω–∏—è –æ–¥–Ω–æ–≥–æ –∏ —Ç–æ–≥–æ –∂–µ –∑–Ω–∞—á–µ–Ω–∏—è: `"benzine"` –∏ `"benzin"`).

–ù–∞ –¥–∞–Ω–Ω–æ–º —ç—Ç–∞–ø–µ **–ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö –Ω–µ –ø—Ä–æ–≤–æ–¥–∏—Ç—Å—è**, —Ç–∞–∫ –∫–∞–∫:

- –õ—é–±–∞—è –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ **–Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–∞**, —Ç–∞–∫ –∫–∞–∫ –æ–Ω–∞ –¥–æ–ª–∂–Ω–∞ –∏–º–∏—Ç–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∞–ª—å–Ω—ã–π –ø–æ—Ç–æ–∫ –Ω–µ–æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.

–ò—Å–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ —É–¥–∞–ª–µ–Ω–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –±—É–¥–µ—Ç –≤—ã–ø–æ–ª–Ω–µ–Ω–æ **–Ω–∞ —ç—Ç–∞–ø–µ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ –ø–∞–π–ø–ª–∞–π–Ω–∞**.

–≠—Ç–∏ –¥—É–±–ª–∏–∫–∞—Ç—ã **–∏–º–µ–µ—Ç —Å–º—ã—Å–ª –æ–±—Ä–∞–±–æ—Ç–∞—Ç—å –∑–∞—Ä–∞–Ω–µ–µ**, —á—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å –ø—Ä–æ–±–ª–µ–º –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏ –º–æ–¥–µ–ª–∏.

### üîπ 4. –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö  
–î–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤ –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü—ã –±—ã–ª–∏ –ø—Ä–∏–≤–µ–¥–µ–Ω—ã –∫ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–º —Ç–∏–ø–∞–º:
- **–î–∞—Ç–∞ –∏ –≤—Ä–µ–º—è** (`date_crawled`, `date_created`, `last_seen`) ‚Üí `datetime64`.
- **–ß–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏** (`price`, `registration_year`, `power`, `kilometer`, `registration_month`, `number_of_pictures`, `postal_code`) ‚Üí `int/float`.
- **–ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏** (`vehicle_type`, `gearbox`, `model`, `fuel_type`, `brand`, `repaired`) ‚Üí `category`.

#### ‚úÖ –≠—Ç–æ –ø–æ–º–æ–≥–∞–µ—Ç:
- –û–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –ø–∞–º—è—Ç—å –ø—Ä–∏ —Ö—Ä–∞–Ω–µ–Ω–∏–∏ –¥–∞–Ω–Ω—ã—Ö.
- –û–±–µ—Å–ø–µ—á–∏—Ç—å –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π.
- –£–ø—Ä–æ—Å—Ç–∏—Ç—å –¥–∞–ª—å–Ω–µ–π—à—É—é –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫—É.

### üî• –ò—Ç–æ–≥
‚úî **–ù–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ —É–Ω–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω—ã**  
‚úî **–í—ã—è–≤–ª–µ–Ω—ã –ø—Ä–æ–ø—É—Å–∫–∏ –∏ –¥—É–±–ª–∏–∫–∞—Ç—ã**  
‚úî **–ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –æ—á–∏—â–µ–Ω—ã –æ—Ç –Ω–µ—è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤**  
‚úî **–°—Ç–æ–ª–±—Ü—ã –ø—Ä–∏–≤–µ–¥–µ–Ω—ã –∫ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–º —Ç–∏–ø–∞–º**  
‚ùó **–ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ–∫–∞ –Ω–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–∞**, —Ç–∞–∫ –∫–∞–∫ —Å–Ω–∞—á–∞–ª–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –¥–∞—Ç–∞—Å–µ—Ç –Ω–∞ **—Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏**. –û—á–∏—Å—Ç–∫–∞ –±—É–¥–µ—Ç –ø—Ä–æ–≤–µ–¥–µ–Ω–∞ –≤ —Ä–∞–º–∫–∞—Ö –ø–∞–π–ø–ª–∞–π–Ω–∞, —á—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å —É—Ç–µ—á–∫–∏ –¥–∞–Ω–Ω—ã—Ö.

–¢–µ–ø–µ—Ä—å –º–æ–∂–Ω–æ –ø–µ—Ä–µ—Ö–æ–¥–∏—Ç—å –∫ **—Ä–∞–∑–¥–µ–ª–µ–Ω–∏—é –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –≤—ã–±–æ—Ä–∫–∏ –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—é –º–æ–¥–µ–ª–∏**. üöÄ

# –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

**–§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–≤–æ–¥–∞ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –≥—Ä–∞—Ñ–∏–∫–æ–≤ –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞**

```python
def perform_exploratory_analysis(df, max_categories=10, drop_columns=None):
    """
    –í—ã–ø–æ–ª–Ω—è–µ—Ç –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö:
    1. –ò—Å–∫–ª—é—á–∞–µ—Ç —É–∫–∞–∑–∞–Ω–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã –∏–∑ –∞–Ω–∞–ª–∏–∑–∞ (–Ω–∞–ø—Ä–∏–º–µ—Ä, —Å—Ç–æ–ª–±—Ü—ã —Å –¥–∞—Ç–∞–º–∏).
    2. –°—Ç—Ä–æ–∏—Ç –≥—Ä–∞—Ñ–∏–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (–≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –∏ –±–æ–∫—Å–ø–ª–æ—Ç—ã).
    3. –°—Ç—Ä–æ–∏—Ç –≥—Ä–∞—Ñ–∏–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã—Ö –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (—Å—Ç–æ–ª–±—á–∞—Ç—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏),
       –ø—Ä–∏ —ç—Ç–æ–º –æ–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ—Ç —á–∏—Å–ª–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–π –¥–æ max_categories.
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        df (pd.DataFrame): –î–∞—Ç–∞—Ñ—Ä–µ–π–º –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞.
        max_categories (int): –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤.
        drop_columns (list): –°–ø–∏—Å–æ–∫ —Å—Ç–æ–ª–±—Ü–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –Ω—É–∂–Ω–æ –∏—Å–∫–ª—é—á–∏—Ç—å –∏–∑ –∞–Ω–∞–ª–∏–∑–∞.
    """

    # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é, –µ—Å–ª–∏ —Å–ø–∏—Å–æ–∫ –Ω–µ –ø–µ—Ä–µ–¥–∞–Ω
    if drop_columns is None:
        drop_columns = ['date_crawled', 'last_seen', 'date_created', 'postal_code', 'number_of_pictures']

    # –ò—Å–∫–ª—é—á–∞–µ–º —É–∫–∞–∑–∞–Ω–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã –∏–∑ –∞–Ω–∞–ª–∏–∑–∞
    df = df.drop(columns=[col for col in drop_columns if col in df.columns], errors='ignore')

    # –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ç–∏–ø–æ–≤ —Å—Ç–æ–ª–±—Ü–æ–≤
    quantitative_cols = df.select_dtypes(include=['float64', 'int64']).columns
    categorical_cols = df.select_dtypes(exclude=['float64', 'int64']).columns

    # –ê–Ω–∞–ª–∏–∑ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã–µ –∏ –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã–µ), –∏—Å–∫–ª—é—á–∞—è —É–¥–∞–ª—ë–Ω–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã
    if len(quantitative_cols) > 0:
        for col in quantitative_cols:
            fig, axes = plt.subplots(1, 2, figsize=(14, 6))

            # –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Å KDE
            sns.histplot(df[col], bins=20, kde=True, ax=axes[0], color='skyblue')
            axes[0].set_title(f"–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ (—Å KDE): {col}")
            axes[0].set_xlabel(col)
            axes[0].set_ylabel("–ß–∞—Å—Ç–æ—Ç–∞")

            # –ë–æ–∫—Å–ø–ª–æ—Ç (Boxplot)
            sns.boxplot(x=df[col], ax=axes[1], color='lightblue')
            axes[1].set_title(f"–ë–æ–∫—Å–ø–ª–æ—Ç: {col}")
            axes[1].set_xlabel(col)

            # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
            plt.tight_layout()
            plt.show()

    # –ê–Ω–∞–ª–∏–∑ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Å –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    if len(categorical_cols) > 0:
        for col in categorical_cols:
            value_counts = df[col].value_counts()

            # –ï—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –±–æ–ª—å—à–µ max_categories, –æ–±—ä–µ–¥–∏–Ω—è–µ–º —Ä–µ–¥–∫–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
            if len(value_counts) > max_categories:
                top_categories = value_counts[:max_categories]
                df[col] = df[col].apply(lambda x: x if x in top_categories.index else '–î—Ä—É–≥–æ–µ')

            plt.figure(figsize=(10, 6))
            sns.countplot(data=df, x=col, palette='Set2', order=df[col].value_counts().index)
            plt.title(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π (–∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫): {col}")
            plt.xlabel(col)
            plt.ylabel("–ß–∞—Å—Ç–æ—Ç–∞")
            plt.xticks(rotation=45)
            plt.show()

# –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∫ –¥–∞–Ω–Ω—ã–º
if autos is not None:
    perform_exploratory_analysis(autos)
else:
    print("‚ùå –û—à–∏–±–∫–∞: –î–∞—Ç–∞—Å–µ—Ç –Ω–µ –∑–∞–≥—Ä—É–∂–µ–Ω, –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å –∞–Ω–∞–ª–∏–∑.")
```

- **–í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ—è–≤–∏–ª–∏—Å—å –¥–æ–≤–æ–ª—å–Ω–æ –≤–∞–∂–Ω—ã–µ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–π —Ä–∞–±–æ—Ç—ã –º–æ–º–µ–Ω—Ç—ã, –∞ –∏–º–µ–Ω–Ω–æ:**
    * –í —Å—Ç–æ–ª–±—Ü–µ `registration_year` –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—Ç —è–≤–Ω—ã–µ –≤—ã–±—Ä–æ—Å—ã(–ì–æ–¥ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ –∞–≤—Ç–æ–º–æ–±–∏–ª—è –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å > 1900, —Ç.–∫ —Ä–∞–Ω—å—à–µ –ø—Ä–æ—Å—Ç–æ –Ω–µ –±—ã–ª–æ –∞–≤—Ç–æ–º–æ–±–∏–ª—è, –∞ –≥–æ–¥–∞ –≤—Ä–æ–¥–µ 4000-–≥–æ –µ—â–µ –Ω–µ –Ω–∞—Å—Ç—É–ø–∏–ª–∏)

## –ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

**–í—ã–≤–µ–ª–∏ –º–∞—Ç—Ä–∏—Ü—É –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è –∏—Å—Ö–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã(–≠—Ç–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ, —Ç.–∫ –≤ —Å–ª—É—á–∞–µ –Ω–∞–ª–∏—á–∏—è –º—É–ª—å—Ç–∏–∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π –º–µ–∂–¥—É –∏—Å—Ö–æ–¥–Ω—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ –≤–µ–ª–∏–∫ —à–∞–Ω—Å –Ω–µ–¥–æ/–ø–µ—Ä–µ–æ–±—É—á–µ–Ω–Ω–æ—Å—Ç–∏ –∏–ª–∏ —É—Ç–µ—á–∫–∏ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞, —á—Ç–æ –æ–¥–Ω–æ–∑–Ω–∞—á–Ω–æ –Ω–µ–≥–∞—Ç–∏–≤–Ω–æ —Å–∫–∞–∂–µ—Ç—Å—è –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–µ –æ–±—É—á–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π)**

**–°—Ç–æ–ª–±—Ü—ã —Ç–∞–±–ª–∏—Ü—ã, —Å–æ–¥–µ—Ä–∂–∞—â–∏–µ –¥–∞—Ç—É –ø—Ä–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ –∫–æ—Ä—Ä–µ–ª—è—Ü—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –Ω–µ —É—á–∏—Ç—ã–≤–∞–µ–º, —Ç.–∫ –æ–Ω–∏ –∏–º–µ—é—Ç —á–µ—Ä–µ–∑–º–µ—Ä–Ω–æ –º–Ω–æ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π**

```python
# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ —Ü–≤–µ—Ç–æ–≤ –¥–ª—è —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
colors = ["blue", "orange", "red"]
bounds = [0.3, 0.5, 0.7, 1.0]
cmap = ListedColormap(colors)

def plot_custom_heatmap(corr_matrix, title):
    """
    –û—Ç–æ–±—Ä–∞–∂–∞–µ—Ç —Ç–µ–ø–ª–æ–≤—É—é –∫–∞—Ä—Ç—É –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        corr_matrix (pd.DataFrame): –ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π.
        title (str): –ó–∞–≥–æ–ª–æ–≤–æ–∫ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã.
    """
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        corr_matrix, annot=True, cmap=cmap, cbar=True, fmt=".2f"
    )
    plt.title(title)
    plt.show()

def identify_continuous_columns(df, quantitative_cols, threshold=10):
    """
    –û–ø—Ä–µ–¥–µ–ª—è–µ—Ç –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        df (pd.DataFrame): –î–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –¥–∞–Ω–Ω—ã–º–∏.
        quantitative_cols (list): –°–ø–∏—Å–æ–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤.
        threshold (int): –ü–æ—Ä–æ–≥ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        list: –°–ø–∏—Å–æ–∫ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.
    """
    return [col for col in quantitative_cols if df[col].nunique() > threshold]

def perform_phik_analysis(df, excluded_columns=None, unique_threshold=10):
    """
    –í—ã–ø–æ–ª–Ω—è–µ—Ç —Ä–∞—Å—á–µ—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã Phik –∏ —Å—Ç—Ä–æ–∏—Ç —Ç–µ–ø–ª–æ–≤—É—é –∫–∞—Ä—Ç—É.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        df (pd.DataFrame): –ò—Å—Ö–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º.
        excluded_columns (list): –°–ø–∏—Å–æ–∫ —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –∏—Å–∫–ª—é—á–µ–Ω–∏—è –∏–∑ –∞–Ω–∞–ª–∏–∑–∞.
        unique_threshold (int): –ü–æ—Ä–æ–≥ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.
    """
    if excluded_columns is None:
        excluded_columns = ["date_crawled", "last_seen", "date_created", "number_of_pictures", "postal_code"]

    # –ò—Å–∫–ª—é—á–∞–µ–º –ø—Ä–æ–±–ª–µ–º–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã
    df = df.drop(columns=[col for col in excluded_columns if col in df.columns], errors="ignore")

    # –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
    quantitative_cols = df.select_dtypes(include=["float64", "int64"]).columns
    continuous_cols = identify_continuous_columns(df, quantitative_cols, threshold=unique_threshold)

    print(f"–ù–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏: {continuous_cols}")

    # –†–∞—Å—á–µ—Ç –º–∞—Ç—Ä–∏—Ü—ã Phik
    phik_matrix = df.phik_matrix(interval_cols=continuous_cols)

    # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
    plot_custom_heatmap(phik_matrix, "Phik Correlation Matrix (–±–µ–∑ –ø—Ä–æ–±–ª–µ–º–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤)")

# –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –∞–Ω–∞–ª–∏–∑–∞
perform_phik_analysis(autos)
```

## –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –ø–µ—Ä–µ–¥ –æ–±—É—á–µ–Ω–∏–µ–º –º–æ–¥–µ–ª–µ–π

**–†–∞–∑–¥–µ–ª–∏–ª–∏ –∏—Å—Ö–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—ã–µ –≤—ã–±–æ—Ä–∫–∏**

```python
def preprocess_and_split_data(
    data,
    target_column="price",
    drop_columns=["postal_code"],
    random_state=42,
    test_size=0.25,
):
    """
    –ü—Ä–µ–¥–æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –¥–∞–Ω–Ω—ã–µ –∏ —Ä–∞–∑–¥–µ–ª—è–µ—Ç –∏—Ö –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫–∏.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        data (pd.DataFrame): –î–∞—Ç–∞—Å–µ—Ç –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏.
        target_column (str): –¶–µ–ª–µ–≤–æ–π —Å—Ç–æ–ª–±–µ—Ü (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 'price').
        drop_columns (list): –°—Ç–æ–ª–±—Ü—ã, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —É–¥–∞–ª–∏—Ç—å –ø–µ—Ä–µ–¥ –æ–±—É—á–µ–Ω–∏–µ–º.
        random_state (int): –ù–∞—á–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª.
        test_size (float): –†–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏.

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        dict: –†–∞–∑–¥–µ–ª—ë–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ (X_train, X_test, y_train, y_test).
    """
    try:
        # –£–¥–∞–ª—è–µ–º –Ω–µ–Ω—É–∂–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã
        data = data.drop(columns=drop_columns, errors="ignore")

        # –†–∞–∑–¥–µ–ª—è–µ–º —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –∏ –ø—Ä–∏–∑–Ω–∞–∫–∏
        X = data.drop(columns=[target_column])
        y = data[target_column]

        # –†–∞–∑–±–∏–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=test_size, random_state=random_state
        )

        # –í—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –≤—ã–±–æ—Ä–∫–∞—Ö
        print(f"‚úÖ –î–∞–Ω–Ω—ã–µ —É—Å–ø–µ—à–Ω–æ —Ä–∞–∑–¥–µ–ª–µ–Ω—ã: \n"
              f"  - X_train: {X_train.shape}, X_test: {X_test.shape} \n"
              f"  - y_train: {y_train.shape}, y_test: {y_test.shape}")

        return {
            "X_train": X_train,
            "X_test": X_test,
            "y_train": y_train,
            "y_test": y_test,
        }

    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞–±–æ—Ç–∫–µ –¥–∞–Ω–Ω—ã—Ö: {e}")
        return None

# –ü—Ä–∏–º–µ–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –∫ —Ç–µ–∫—É—â–µ–º—É –¥–∞—Ç–∞—Å–µ—Ç—É
split_data = preprocess_and_split_data(autos)
```

**–ò–∑–±–∞–≤–∏–ª–∏—Å—å –æ—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ**

```python
def process_duplicates_and_missing_values(df, df_name):
    """
    –£–¥–∞–ª—è–µ—Ç –¥—É–±–ª–∏–∫–∞—Ç—ã –∏ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        df (pd.DataFrame): –ò—Å—Ö–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º.
        df_name (str): –ù–∞–∑–≤–∞–Ω–∏–µ –¥–∞—Ç–∞—Å–µ—Ç–∞ (–Ω–∞–ø—Ä–∏–º–µ—Ä, "X_train" –∏–ª–∏ "X_valid").

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        pd.DataFrame: –û—á–∏—â–µ–Ω–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç.
    """
    try:
        if df is None or df.empty:
            print(f"‚ö† –î–∞—Ç–∞—Å–µ—Ç {df_name} –ø—É—Å—Ç –∏–ª–∏ –Ω–µ –∑–∞–≥—Ä—É–∂–µ–Ω!")
            return df

        # –ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ–º –¥—É–±–ª–∏–∫–∞—Ç—ã
        duplicates_count = df.duplicated().sum()
        if duplicates_count > 0:
            print(f"‚ö† –í {df_name} –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ {duplicates_count} –¥—É–±–ª–∏–∫–∞—Ç–æ–≤. –£–¥–∞–ª—è–µ–º...")
            df = df.drop_duplicates().reset_index(drop=True)

        # –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤
        missing_values_count = df.isna().sum().sum()
        if missing_values_count > 0:
            print(f"‚ö† –í {df_name} –Ω–∞–π–¥–µ–Ω–æ {missing_values_count} –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π. –£–¥–∞–ª—è–µ–º...")
            df = df.dropna().reset_index(drop=True)

        print(f"‚úÖ {df_name} –æ—á–∏—â–µ–Ω. –ù–æ–≤—ã–π —Ä–∞–∑–º–µ—Ä: {df.shape}")

        return df

    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –æ—á–∏—Å—Ç–∫–µ {df_name}: {e}")
        return df


def clean_training_data(X_train, y_train):
    """
    –£–¥–∞–ª—è–µ—Ç –¥—É–±–ª–∏–∫–∞—Ç—ã, –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –ø—Ä–æ–ø—É—Å–∫–∏ –∏ —Å–∏–Ω—Ö—Ä–æ–Ω–∏–∑–∏—Ä—É–µ—Ç y_train —Å X_train.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        X_train (pd.DataFrame): –ü—Ä–∏–∑–Ω–∞–∫–∏.
        y_train (pd.Series): –¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è.

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        tuple: –û—á–∏—â–µ–Ω–Ω—ã–µ X_train –∏ y_train.
    """
    try:
        if X_train is None or y_train is None:
            print("‚ö† –û—à–∏–±–∫–∞: –û–¥–∏–Ω –∏–∑ –≤—Ö–æ–¥–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ None!")
            return None, None

        if X_train.empty or y_train.empty:
            print("‚ö† –û—à–∏–±–∫–∞: –û–¥–∏–Ω –∏–∑ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞—Ç–∞—Å–µ—Ç–æ–≤ –ø—É—Å—Ç!")
            return X_train, y_train

        # –û–±—ä–µ–¥–∏–Ω—è–µ–º X_train –∏ y_train –¥–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–≥–æ —É–¥–∞–ª–µ–Ω–∏—è —Å—Ç—Ä–æ–∫
        df_combined = X_train.copy()
        df_combined["target"] = y_train.values  # –î–æ–±–∞–≤–ª—è–µ–º —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é

        # –ß–∏—Å—Ç–∏–º –¥–∞–Ω–Ω—ã–µ
        df_cleaned = process_duplicates_and_missing_values(df_combined, "X_train")

        # –†–∞–∑–¥–µ–ª—è–µ–º –æ–±—Ä–∞—Ç–Ω–æ
        X_train_cleaned = df_cleaned.drop(columns=["target"])
        y_train_cleaned = df_cleaned["target"]

        print(f"‚úÖ –†–∞–∑–º–µ—Ä—ã X_train –ø–æ—Å–ª–µ –æ—á–∏—Å—Ç–∫–∏: {X_train_cleaned.shape}")
        print(f"‚úÖ –†–∞–∑–º–µ—Ä—ã y_train –ø–æ—Å–ª–µ –æ—á–∏—Å—Ç–∫–∏: {y_train_cleaned.shape}")

        return X_train_cleaned, y_train_cleaned

    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –æ—á–∏—Å—Ç–∫–µ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: {e}")
        return None, None


# –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ª–∏—á–∏–µ –¥–∞–Ω–Ω—ã—Ö –ø–µ—Ä–µ–¥ –æ–±—Ä–∞–±–æ—Ç–∫–æ–π
if isinstance(split_data, dict) and "X_train" in split_data and "y_train" in split_data:
    X_train_cleaned, y_train_cleaned = clean_training_data(split_data["X_train"], split_data["y_train"])
else:
    print("‚ùå –û—à–∏–±–∫–∞: split_data –Ω–µ —Å–æ–¥–µ—Ä–∂–∏—Ç X_train –∏–ª–∏ y_train!")
```

**–¢.–∫ –ø—Ä–æ–ø—É—Å–∫–∏ —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç –Ω–µ –±–æ–ª–µ–µ 10% —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –ø—Ä–æ—Å—Ç–æ —É–¥–∞–ª–∏–ª–∏ –∏—Ö**

## –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

**–ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä–æ–≤–∞–ª–∏ –ø–∞–π–ø–ª–∞–π–Ω –∏ –æ–±—É—á–∏–ª–∏ –º–æ–¥–µ–ª–∏ —Å –ø–æ–¥–±–æ—Ä–æ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤**

```python
# üõ† –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ RMSE
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# üîç **–û–ø—Ä–µ–¥–µ–ª—è–µ–º —á–∏—Å–ª–æ–≤—ã–µ –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏**
num_columns = [col for col in ['registration_year', 'power', 'kilometer'] if col in X_train_cleaned.columns]
cat_columns = [col for col in ['vehicle_type', 'gearbox', 'model', 'fuel_type', 'repaired', 'registration_month', 'number_of_pictures'] if col in X_train_cleaned.columns]

# –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
transformers_list = []

if num_columns:
    transformers_list.append(('num_scaling', Pipeline([
        ('num_imputer', SimpleImputer(strategy='median')),
        ('num_scaler', StandardScaler())
    ]), num_columns))

if cat_columns:
    transformers_list.append(('cat_encoding', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), cat_columns))

# –ï—Å–ª–∏ –Ω–µ—Ç –Ω–∏ –æ–¥–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞, –±—É–¥–µ—Ç –æ—à–∏–±–∫–∞
if not transformers_list:
    raise ValueError("‚ùå –û—à–∏–±–∫–∞: –ù–µ—Ç –Ω–∏ —á–∏—Å–ª–æ–≤—ã—Ö, –Ω–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏!")

# –°–æ–∑–¥–∞–µ–º –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç—á–∏–∫ —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π
preprocessor = ColumnTransformer(transformers=transformers_list)

# üîπ **–û–ø—Ä–µ–¥–µ–ª—è–µ–º –º–æ–¥–µ–ª–∏ –∏ —Ä–∞—Å—à–∏—Ä–µ–Ω–Ω—ã–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã**
param_distributions = [
    {
        'model': [Ridge()],
        'model__alpha': np.logspace(-2, 2, 10)
    },
    {
        'model': [LGBMRegressor()],
        'model__n_estimators': [100, 300, 500, 1000],
        'model__learning_rate': [0.01, 0.05, 0.1, 0.2],
        'model__max_depth': [3, 5, 7, 10]
    },
    {
        'model': [CatBoostRegressor(verbose=0)],
        'model__iterations': [100, 300, 500, 1000],
        'model__learning_rate': [0.01, 0.05, 0.1, 0.2],
        'model__depth': [3, 5, 7, 10]
    }
]

# üîπ **–ö–æ–Ω–≤–µ–π–µ—Ä –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏**
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', Ridge())  
])

# üîπ **RandomizedSearchCV —Å —É–≤–µ–ª–∏—á–µ–Ω–Ω—ã–º —á–∏—Å–ª–æ–º –∏—Ç–µ—Ä–∞—Ü–∏–π**
random_search = RandomizedSearchCV(
    pipeline,
    param_distributions=param_distributions,
    n_iter=15,
    scoring='neg_root_mean_squared_error',
    cv=3,
    n_jobs=-1,
    verbose=2,
    return_train_score=True,
    random_state=42
)

# üîπ **–û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π —Å –∑–∞–º–µ—Ä–æ–º –≤—Ä–µ–º–µ–Ω–∏**
try:
    start_time = time.time()
    random_search.fit(X_train_cleaned, y_train_cleaned)
    training_time = time.time() - start_time
    best_model = random_search.best_estimator_

    print("\nüöÄ –õ—É—á—à–∞—è –º–æ–¥–µ–ª—å:")
    print(best_model)

    # –ò–∑–≤–ª–µ–∫–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏
    cv_results_df = pd.DataFrame(random_search.cv_results_)

    # –î–æ–±–∞–≤–ª—è–µ–º —Å—Ç–æ–ª–±–µ—Ü —Å –Ω–∞–∑–≤–∞–Ω–∏–µ–º –º–æ–¥–µ–ª–∏
    cv_results_df['–ú–æ–¥–µ–ª—å'] = cv_results_df['param_model'].apply(lambda x: x.__class__.__name__)

    # –í—ã–±–∏—Ä–∞–µ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –º–æ–¥–µ–ª–∏
    param_columns = ['–ú–æ–¥–µ–ª—å']
    if 'param_model__alpha' in cv_results_df.columns:
        param_columns.append('param_model__alpha')
    if 'param_model__n_estimators' in cv_results_df.columns:
        param_columns.append('param_model__n_estimators')
    if 'param_model__learning_rate' in cv_results_df.columns:
        param_columns.append('param_model__learning_rate')
    if 'param_model__max_depth' in cv_results_df.columns:
        param_columns.append('param_model__max_depth')
    if 'param_model__iterations' in cv_results_df.columns:
        param_columns.append('param_model__iterations')

    # –û—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ –Ω—É–∂–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã
    cv_results_df = cv_results_df[param_columns + ['mean_test_score', 'mean_train_score', 'mean_fit_time', 'mean_score_time']]

    # –ò–Ω–≤–µ—Ä—Ç–∏—Ä—É–µ–º –º–µ—Ç—Ä–∏–∫—É RMSE (–∏–∑ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π)
    cv_results_df['mean_test_score'] = -cv_results_df['mean_test_score']
    cv_results_df['mean_train_score'] = -cv_results_df['mean_train_score']

    # –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã
    cv_results_df.rename(columns={
        'mean_test_score': '–°—Ä–µ–¥–Ω–∏–π RMSE (–≤–∞–ª–∏–¥–∞—Ü–∏—è)',
        'mean_train_score': '–°—Ä–µ–¥–Ω–∏–π RMSE (–æ–±—É—á–µ–Ω–∏–µ)',
        'mean_fit_time': '–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è (—Å–µ–∫)',
        'mean_score_time': '–í—Ä–µ–º—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è (—Å–µ–∫)',
        'param_model__alpha': 'Alpha (Ridge)',
        'param_model__n_estimators': '–ö–æ–ª-–≤–æ –¥–µ—Ä–µ–≤—å–µ–≤ (GBM)',
        'param_model__learning_rate': 'LR (GBM)',
        'param_model__max_depth': '–ì–ª—É–±–∏–Ω–∞ –¥–µ—Ä–µ–≤—å–µ–≤ (GBM)',
        'param_model__iterations': '–ö–æ–ª-–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π (CatBoost)'
    }, inplace=True)

    # üîπ **–°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ –º–µ—Ç—Ä–∏–∫–µ RMSE (–≤–∞–ª–∏–¥–∞—Ü–∏—è)**
    cv_results_df = cv_results_df.sort_values(by="–°—Ä–µ–¥–Ω–∏–π RMSE (–≤–∞–ª–∏–¥–∞—Ü–∏—è)")

    # üîπ **–°–æ—Ö—Ä–∞–Ω—è–µ–º –∏ –≤—ã–≤–æ–¥–∏–º —Ç–∞–±–ª–∏—Ü—É —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏**
    cv_results_df.to_csv("cv_results_sorted.csv", index=False)
    final_display("\nüìä –¢–∞–±–ª–∏—Ü–∞ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ (–æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–∞—è) —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∞ –≤ 'cv_results_sorted.csv'.")
    display(cv_results_df)

    # üîπ **–ó–∞–º–µ—Ä—è–µ–º –≤—Ä–µ–º—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è**
    start_pred_time = time.time()
    _ = best_model.predict(X_train_cleaned[:1000])
    pred_time = (time.time() - start_pred_time) / 1000  

    display(f"\n‚úÖ RMSE (–∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è): {-random_search.best_score_:.4f}")
    display(f"‚è≥ –í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è: {training_time:.2f} —Å–µ–∫.")
    (f"‚ö° –°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è: {pred_time:.6f} —Å–µ–∫. –Ω–∞ –æ–¥–∏–Ω –ø—Ä–∏–º–µ—Ä")

    # üî• **–§–ò–ù–ê–õ–¨–ù–û–ï –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï –Ω–∞ X_test**
    y_test_pred = best_model.predict(X_test)
    test_rmse = rmse(y_test, y_test_pred)

    title_display(f"\nüèÅ –ò—Ç–æ–≥–æ–≤—ã–π RMSE –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {test_rmse:.4f}")

    # üîπ **–ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏**
    if test_rmse < 2500:
        title_display("\nüéØ –ú–æ–¥–µ–ª—å —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –∫—Ä–∏—Ç–µ—Ä–∏—é RMSE < 2500!")
    else:
        err_display("\n‚ö†Ô∏è –ú–æ–¥–µ–ª—å –Ω–µ –ø—Ä–æ—à–ª–∞ –∫—Ä–∏—Ç–µ—Ä–∏–π RMSE. –¢—Ä–µ–±—É–µ—Ç—Å—è —É–ª—É—á—à–µ–Ω–∏–µ.")


except Exception as e:
    err_display(f"\n‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏ –º–æ–¥–µ–ª–∏: {e}")
```

**–ù–∞–∏–±–æ–ª–µ–µ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ–π –¥–ª—è –∑–∞–∫–∞–∑—á–∏–∫–∞ —è–≤–ª—è–µ—Ç—Å—è 5 –º–æ–¥–µ–ª—å(–û–Ω–∞ –∏–º–µ–µ—Ç –æ—Ç–ª–∏—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–ª—é—á–µ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ –∏ –ø—Ä–∏ —ç—Ç–æ–º –±—ã—Å—Ç—Ä–æ –æ–±—É—á–∞–µ—Ç—Å—è –∏ –¥–µ–ª–∞–µ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è)**

**–î–ª—è –±–æ–ª—å—à–µ–π –æ–±—ä–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ—Ü–µ–Ω–∫–∏ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ —Ä–∞–±–æ—Ç—ã –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ –≤—ã–≤–µ–ª–∏ –≥—Ä–∞—Ñ–∏–∫–∏ –∞–Ω–∞–ª–∏–∑–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤ –º–æ–¥–µ–ª–∏ –∏ SHAP-–≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤**

```python
def train_and_analyze_residuals_with_shap(best_model, X_train, X_test, y_train, y_test):
    """
    –ê–Ω–∞–ª–∏–∑ –æ—Å—Ç–∞—Ç–∫–æ–≤ –º–æ–¥–µ–ª–∏ –∏ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º SHAP.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        best_model: –û–ø—Ç–∏–º–∞–ª—å–Ω–∞—è –º–æ–¥–µ–ª—å (LGBMRegressor —Å n_estimators=300, learning_rate=0.2, max_depth=7).
        X_train, X_test (pd.DataFrame): –ü—Ä–∏–∑–Ω–∞–∫–∏ –æ–±—É—á–∞—é—â–µ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏.
        y_train, y_test (pd.Series): –¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –æ–±—É—á–∞—é—â–µ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏.
    """

    display("üöÄ –°—Ç–∞—Ä—Ç –∞–Ω–∞–ª–∏–∑–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤ –∏ SHAP")

    # üîπ **–ó–∞–º–µ–Ω—è–µ–º –º–æ–¥–µ–ª—å –Ω–∞ –æ–ø—Ç–∏–º–∞–ª—å–Ω—É—é (–º–æ–¥–µ–ª—å ‚Ññ5)**
    best_model.named_steps['model'] = LGBMRegressor(n_estimators=300, learning_rate=0.2, max_depth=7)

    # üîπ **–ò–∑–≤–ª–µ–∫–∞–µ–º –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç—á–∏–∫ –∏ –º–æ–¥–µ–ª—å**
    preprocessor = best_model.named_steps['preprocessor']
    true_model = best_model.named_steps['model']

    # üîπ **–ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞–Ω–Ω—ã–µ**
    X_train_preprocessed = preprocessor.transform(X_train)
    X_test_preprocessed = preprocessor.transform(X_test)
    feature_names = preprocessor.get_feature_names_out()

    X_train_preprocessed = pd.DataFrame(X_train_preprocessed, columns=feature_names)
    X_test_preprocessed = pd.DataFrame(X_test_preprocessed, columns=feature_names)

    # üîπ **–û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å ‚Ññ5 –Ω–∞ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö**
    true_model.fit(X_train_preprocessed, y_train)

    # üîπ **–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∏ —Ä–∞—Å—á—ë—Ç –æ—Å—Ç–∞—Ç–∫–æ–≤**
    y_train_pred = true_model.predict(X_train_preprocessed)
    y_test_pred = true_model.predict(X_test_preprocessed)
    residuals = y_test - y_test_pred

    # üîπ **–ì—Ä–∞—Ñ–∏–∫–∏ –∞–Ω–∞–ª–∏–∑–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤**
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    fig.suptitle("–ê–Ω–∞–ª–∏–∑ –æ—Å—Ç–∞—Ç–∫–æ–≤", fontsize=14)

    # –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤
    sns.histplot(residuals, kde=True, bins=30, color="blue", ax=axes[0])
    axes[0].set_title("–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤")
    axes[0].set_xlabel("–û—Å—Ç–∞—Ç–∫–∏")
    axes[0].set_ylabel("–ß–∞—Å—Ç–æ—Ç–∞")
    axes[0].grid(True)

    # –û—Å—Ç–∞—Ç–∫–∏ vs –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    sns.scatterplot(x=y_test_pred, y=residuals, ax=axes[1], color="red")
    axes[1].axhline(0, color="black", linestyle="--")
    axes[1].set_title("–û—Å—Ç–∞—Ç–∫–∏ vs –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è")
    axes[1].set_xlabel("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è")
    axes[1].set_ylabel("–û—Å—Ç–∞—Ç–∫–∏")
    axes[1].grid(True)

    # Q-Q Plot
    stats.probplot(residuals, dist="norm", plot=axes[2])
    axes[2].set_title("Q-Q Plot")

    plt.tight_layout()
    plt.show()

    display("‚úÖ –ê–Ω–∞–ª–∏–∑ –æ—Å—Ç–∞—Ç–∫–æ–≤ –∑–∞–≤–µ—Ä—à–µ–Ω.")

    # üîπ **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤**
    display("\nüìä –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤:")
    display(f"–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤: {np.mean(residuals):.4f}")
    display(f"–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤: {np.std(residuals):.4f}")
    display(f"–°–∫–æ—à–µ–Ω–Ω–æ—Å—Ç—å –æ—Å—Ç–∞—Ç–∫–æ–≤: {pd.Series(residuals).skew():.4f}")
    display(f"–≠–∫—Å—Ü–µ—Å—Å –æ—Å—Ç–∞—Ç–∫–æ–≤: {pd.Series(residuals).kurt():.4f}")

    # üîπ **SHAP-–∞–Ω–∞–ª–∏–∑**
    display("\nüîç –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è –º–æ–¥–µ–ª–∏ —Å –ø–æ–º–æ—â—å—é SHAP:")
    
    try:
        # –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–º–µ—Ä–æ–≤ –¥–ª—è SHAP (–Ω–∞–ø—Ä–∏–º–µ—Ä, 500)
        shap_sample = X_test_preprocessed.sample(min(500, len(X_test_preprocessed)))

        # SHAP-–∞–Ω–∞–ª–∏–∑ –¥–ª—è –¥–µ—Ä–µ–≤—å–µ–≤ —Ä–µ—à–µ–Ω–∏–π (LightGBM)
        explainer = shap.TreeExplainer(true_model)
        shap_values = explainer(shap_sample)

        # SHAP Beeswarm Plot
        plt.figure(figsize=(10, 6))
        shap.summary_plot(shap_values, shap_sample, feature_names=feature_names)
        display("‚úÖ SHAP-–≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è (–ø—á–µ–ª–∏–Ω—ã–π —Ä–æ–π) –∑–∞–≤–µ—Ä—à–µ–Ω–∞.")

        # SHAP Bar Plot
        plt.figure(figsize=(10, 6))
        shap.summary_plot(shap_values, shap_sample, feature_names=feature_names, plot_type="bar")
        display("‚úÖ SHAP-–≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è (–±–∞—Ä-–≥—Ä–∞—Ñ–∏–∫) –∑–∞–≤–µ—Ä—à–µ–Ω–∞.")

        # SHAP Waterfall Plot (–¥–ª—è –æ–¥–Ω–æ–≥–æ –ø—Ä–∏–º–µ—Ä–∞)
        shap.waterfall_plot(shap_values[0])
        display("‚úÖ SHAP-–≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è (–≤–æ–¥–æ–ø–∞–¥–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫) –∑–∞–≤–µ—Ä—à–µ–Ω–∞.")

    except Exception as e:
        display(f"‚ö† –û—à–∏–±–∫–∞ –ø—Ä–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ SHAP: {e}")

    display("\nüéØ –ê–Ω–∞–ª–∏–∑ –∑–∞–≤–µ—Ä—à—ë–Ω.")
```

```python
# üî• **–ó–∞–ø—É—Å–∫ –∞–Ω–∞–ª–∏–∑–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤ –∏ SHAP**
train_and_analyze_residuals_with_shap(
    best_model, 
    X_train_cleaned, 
    split_data["X_test"], 
    y_train_cleaned, 
    split_data["y_test"]
)
```

## üöÄ **–í—ã–≤–æ–¥ –ø–æ –æ–±—É—á–µ–Ω–∏—é –º–æ–¥–µ–ª–µ–π –∏ –∞–Ω–∞–ª–∏–∑—É –∏—Ö –∫–∞—á–µ—Å—Ç–≤–∞**
